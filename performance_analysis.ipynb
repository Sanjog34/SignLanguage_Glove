```python
import pandas as pd

df=pd.read_csv("sensor_data_8flex_new_ka.csv")
```

```python
df.head()
```

```python
df.count
```

```python
ka_df = df[df["label"] == "ka"]
kha_df = df[df["label"] == "kha"]
ga_df = df[df["label"] == "ga"]
gha_df = df[df["label"] == "gha"]
ek_df = df[df["label"] == "ek"]
dui_df = df[df["label"] == "dui"]
tin_df = df[df["label"] == "teen"]
```

```python
ka_df.count
```

```python
features = ["idxUp","idxLow","midUp","midLow","ringUp","ringLow","thumb","pinky"]
ka_features = ka_df[features]
kha_features = kha_df[features]
ga_features = ga_df[features]
gha_features = gha_df[features]
ek_features = ek_df[features]
dui_features = dui_df[features]
tin_features = tin_df[features]
```

```python
ka_features.head
```

```python
corr_ka = ka_features.corr(method="pearson")
corr_kha = kha_features.corr(method="pearson")
corr_ga = ga_features.corr(method="pearson")
corr_gha = gha_features.corr(method="pearson")
corr_ek = ek_features.corr(method="pearson")
corr_dui = dui_features.corr(method="pearson")
corr_tin = tin_features.corr(method="pearson")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(corr_ka, annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Matrix for Gesture 'ka'")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_kha, annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Matrix for Gesture 'kha'")
plt.show()
```
