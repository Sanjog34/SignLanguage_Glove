```python
import pandas as pd
```

```python
df=pd.read_csv("sensor_data_8flex_alphaNEP.csv")
```

```python
import matplotlib.pyplot as plt

fist_data = df[df["label"] == "ek"]

feature_cols = ["idxUp","idxLow","midUp","midLow",
                "ringUp","ringLow","thumb","pinky"]

plt.figure(figsize=(12, 14))
```

```python
for i, feature in enumerate(feature_cols, 1):
    plt.subplot(4, 2, i)
    plt.plot(fist_data[feature].values)
    plt.title(feature)
    plt.xlabel("Sample index")
    plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

```python



# Filter label
fist_data = df[df["label"] == "ga"]

# Feature columns
features = ["idxUp","idxLow","midUp","midLow",
            "ringUp","ringLow","thumb","pinky"]

# Compute mean values of features
mean_values = fist_data[features].mean().values  # <= convert to numpy array

plt.figure(figsize=(10, 5))

plt.plot(features, mean_values, marker="o")  # now both x & y are valid

plt.title("Average Feature Values for Label 'ga'")
plt.xlabel("Features")
plt.ylabel("Value")

plt.grid(True)
plt.show()
```
