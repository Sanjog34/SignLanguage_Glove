```python
import pandas as pd
```

```python
df=pd.read_csv("sensor_data_8flex.csv")
```

```python
import matplotlib.pyplot as plt

fist_data = df[df["label"] == "two"]

feature_cols = ["idxUp","idxLow","midUp","midLow",
                "ringUp","ringLow","thumb","pinky"]

plt.figure(figsize=(12, 14))
```

```python
for i, feature in enumerate(feature_cols, 1):
    plt.subplot(4, 2, i)
    plt.plot(fist_data[feature].values)
    plt.title(feature)
    plt.xlabel("Sample index")
    plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

```python



# Filter label
fist_data = df[df["label"] == "two"]

# Feature columns
features = ["idxUp","idxLow","midUp","midLow",
            "ringUp","ringLow","thumb","pinky"]

# Compute mean values of features
mean_values = fist_data[features].mean().values  # <= convert to numpy array

plt.figure(figsize=(10, 5))

plt.plot(features, mean_values, marker="o")  # now both x & y are valid

plt.title("Average Feature Values for Label 'two'")
plt.xlabel("Features")
plt.ylabel("Value")

plt.grid(True)
plt.show()
```

```python


# Filter label
thumbup_data = df[df["label"] == "thumbup"]
thumbup_data = thumbup_data[thumbup_data["midUp"] >= 130]
# Feature columns
features = ["idxUp","idxLow","midUp","midLow",
            "ringUp","ringLow","thumb","pinky"]

plt.figure(figsize=(12, 6))

colors = plt.cm.tab10.colors  # 10 distinct colors

# Plot each sample
for i, row in thumbup_data.iterrows():
    plt.plot(features, row[features].values, marker='o', alpha=0.5)

plt.title("Feature Values for All Samples of Label 'thumbup'")
plt.xlabel("Features")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
